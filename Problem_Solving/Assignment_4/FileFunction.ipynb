- The final output should be the attached example file. One is for yes and zero means no (.5pts).
- Chaining the analysis as pipeline 1.5pts.

**My Thoughts**:

```python
inPath = r"C:\Users\tiffa\OneDrive\Desktop\Masters in Bioinformatics\Problem Solving\Homeworks\FGFR_fusion.fasta"
#path for the fusion fasta file copied directly from my directory
#I need to make my path a raw literal string or else it throws errors
outPath = r"C:\Users\tiffa\OneDrive\Desktop\Masters in Bioinformatics\Problem Solving\Homeworks\FilesHWout.txt"
```

```python
def function_A(input_file):
    outFile = outPath #create file for writing
    out = open(outFile, 'w')
    #define my out file object as a raw literal string, or else it throws errors
    with open(input_file, 'r') as f_input:
        for line in f_input:
            #for each line in our read in file.
            if 'fibroblast growth factor receptor 1 isoform' in line:
                #if the line has the key words, we will record it and it's sequence
                while not line.startswith('\n'):
                    #I am using starting with a newline as a delimeter
                    #so it should write all lines up to an empty line
                    out.write(line)
                    line = next(f_input)
                    #set the line to be the next line
    out.close()
    #remember to close out the file when you're done writing!
    function_B(outFile)
```

```python

#I am going to keep a dictionary to store my fasta sequences. 
def function_B(output_file):

    fastaSeq = {} #initialize my empty dictionary
    #this sequence will populate my dictionary for easier data analysis

    with open(output_file, 'r') as f:
        #open and read my file
        header = ''#initialize empty header variable
        sequence = ''#initialize empty sequence variable
        
        for line in f:#iterate over each line in the file
            
            #line = line.strip()#clean up the line of white spaces and newline chars
            #it's position 563 with the newline characters
            if line.startswith('>'):
                if header:  # If header is not empty (i.e., not the first header)
                    fastaSeq[header] = sequence  # Store the previous header and sequence
                header = line[0:]  # Extract header without '>'
                sequence = ''  # Reset sequence for the new header
            else:
                sequence += line  # Concatenate lines to form the sequence
        # Store the last header and sequence
        if header:
            fastaSeq[header] = sequence

    #now the dictionary should be set, we can go through our dictionary, open the file as write
    out = open(output_file, 'w')#open the file as writing mode
    out.write('ID\tValine\tIsoleucine\tGlutamic Acid\n')
    #put a header into our out file
    for header, sequence in fastaSeq.items(): #for each pair in our dictionary
            header = header.strip() #clean up the header of spaces and blank characters before writing
            out.write(header) #write the header
            AA = sequence[563] #check position 564, python is 0 index based, so it would be index 563
            if AA == 'V':
                 #if it is a V, put a 1 in the first position
                 out.write(' 1\t0\t0\n')

            elif AA == 'I':
                 #if it is an I, put a 1 in the second position
                 out.write(' 0\t1\t0\n')

            elif AA == 'E':
                #if it is an E, put a 1 in the third position
                out.write(' 0\t0\t1\n')

            else:
                #put 0 for all positions
                out.write(' 0\t0\t0\n')
        
```
