```r
#sample 30 males and 30 females
# first I will split the wh by males and females

wh_males <- wh[wh$Gender == "Male", ]#this creates a dataframe of just males
wh_females <- wh[wh$Gender == "Female", ]#this creates a dataframe of just males

# Sample 30 males from the data set
# sample from 1 through all the rows of wh_males, select 30 without replacement
sample_males <- wh_males[sample(1:nrow(wh_males), 30, replace=FALSE), ]

# Sample 30 females from the data set
sample_females <- wh_females[sample(1:nrow(wh_females), 30, replace=FALSE), ]
```

```r
library("bootstrap") #import library

#for jackknife function, need to compute mean
# So susbsitute the theta for the mean function 

#create a mean function, where my arguments are data and indicies
# it computes the mean of the subset of data
mean_function <- function(data, indicies){
    return(mean(data[indicies]))
}

#put in the data and the function into the jackknife function
#Perform jackknife for males
# jackknife function is applied to columns in the sample_males data set, 
# using the mean_function as the function to compute the jackknife estimate
jackknife_males_height <- jackknife(sample_males$Height, mean_function)
jackknife_males_weight <- jackknife(sample_males$Weight, mean_function)


print("Jackknife estimate for the males mean, height then weight:")
print(mean(jackknife_males_height[[3]]))#this will give the estimate of theta, 3rd element is theta
print(mean(jackknife_males_weight[[3]]))
cat("\n")

# Perform jackknife for females
jackknife_females_height <- jackknife(sample_females$Height, mean_function)
jackknife_females_weight <- jackknife(sample_females$Weight, mean_function)
# Print out in the same manner
print("Jackknife estimate for the females mean, height then weight:")
print(mean(jackknife_females_height[[3]])) 
print(mean(jackknife_females_weight[[3]]))
cat("\n")

# Combine male and female data for the combined group
sample_combined <- rbind(sample_males, sample_females)

# Perform jackknife for combined group
jackknife_combined_height <- jackknife(sample_combined$Height, mean_function)
jackknife_combined_weight <- jackknife(sample_combined$Weight, mean_function)
# Print out in the same manner 
print("Jackknife estimate for the combined group mean, height then weight:")
print(mean(jackknife_combined_height[[3]]))
print(mean(jackknife_combined_weight[[3]])) # This gives the jackknife mean estimate
```

```r
# I am going to repeat the same thing, but instead of mean
# I will replace theta with a variance function. 

# Variance
variance_function <- function(data, indices) {
  return(var(data[indices]))  # Calculate variance for the selected indices
}

# Perform jackknife for males
# same as above, except perform jackknife using variance as the function
jackknife_males_height <- jackknife(sample_males$Height, variance_function)
jackknife_males_weight <- jackknife(sample_males$Weight, variance_function)

# Jackknife estimate of variance for males
print("Jackknife estimate for the males variance, height then weight variance:")
print(mean(jackknife_males_height[[3]]))  # This gives the jackknife estimate of variance for height
print(mean(jackknife_males_weight[[3]]))  # This gives the jackknife estimate of variance for weight
cat("\n")

# Perform jackknife for females
jackknife_females_height <- jackknife(sample_females$Height, variance_function)
jackknife_females_weight <- jackknife(sample_females$Weight, variance_function)

# Jackknife estimate of variance for females
print("Jackknife estimate for the females variance, height then weight variance:")
print(mean(jackknife_females_height[[3]]))  # Jackknife estimate for height variance
print(mean(jackknife_females_weight[[3]]))  # Jackknife estimate for weight variance
cat("\n")

# Combine male and female data for the combined group
sample_combined <- rbind(sample_males, sample_females)

# Perform jackknife for combined group
jackknife_combined_height <- jackknife(sample_combined$Height, variance_function)
jackknife_combined_weight <- jackknife(sample_combined$Weight, variance_function)

# Jackknife estimate of variance for combined group
print("Jackknife estimate for the combined group variance, height then weight variance:")
print(mean(jackknife_combined_height[[3]]))  # Jackknife estimate for combined height variance
print(mean(jackknife_combined_weight[[3]]))  # Jackknife estimate for combined weight variance
```

**Using the sample function in R, write a bootstrap routine that outputs the bootstrap mean. Explain the logic by annotating the code.**

```r
# Bootstrap mean function
# Bootstrap resamples the data with replacement
# Need to pull a sample the same size as the original dataset, 

# Define the function, inputs will be data, and # of times you want to resample
bootstrap_mean <- function(data, num_resamples){
    means_v <- numeric(num_resamples) #initializes the vector of all your means of your resamples

    # Resampling process
    for (i in 1: num_resamples){ # Will resample num_resamples times
        # Resample the number of times the length of the dataset
        resample <- sample( data, size = length(data), replace = TRUE) # With replacement!!!
        #boot strap samples with replacement!!

        # Since we want bootstrap with mean
        # calculate the mean of the resampled data
        means_v[i] <- mean(resample) # Add the mean into the means vector at the correct index
    }
    return( mean(means_v)) #return the mean of the means
}
```

```r
# Let's test with the sample sets from the previous question
bootstrap_mean(sample_males$Height, 1000)
#output looks good!
```
